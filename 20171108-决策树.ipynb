```r
data<-read.csv("all.csv",header=TRUE,sep=",");
```

```r
data
```

```r
data<-cbind(data,rep(0,length(data[,1])));
```

```r
data[which(data[,9]>24),ncol(data)]<-1;
data<-as.matrix(data);
dimnames(data)[2]<-list(c('Sample_geo_accession','gender','dagnoss','grade','age','RT','CT','event','survival','survivalType','codel','IDH1','HDType'));
data<-as.data.frame(data);
```

```r
data
```

最后一列为生存时间<24 为0 高风险， >24 为1 低风险。

```r
library(rpart)
library(rpart.plot)
library(extrafont)
options(repr.plot.width=12, repr.plot.height=9)
```

```r
ct <- rpart.control(xval=10, minsplit=1, cp=0.1)
```

```r
##grade是分类属性即要分的类型（eg最后的结果是否是癌症这样）；~后面是可以得到结果的属性；data是导入的表格；class是离散型；
fit <- rpart(dagnoss ~ RT + CT + grade + gender,data=data, method="class",control=ct,
             parms = list(split = "gini"),);
```

```r
fit
```

```r
rpart.plot(fit, branch=1, branch.type=2, type=1,extra=102, box.col="pink", border.col="black", split.col="red", split.cex=1.2, main="Kyphosis决策树",family='Helvetica',font=5);
```

```r
fit <- rpart(HDType ~ RT + CT,data=data, method="class",control=ct,
             parms = list(split = "gini"),);
```

```r
fit
```

```r
rpart.plot(fit, branch=1, branch.type=2, type=1,extra=102, box.col="yellow", border.col="black", split.col="red", split.cex=1.2, main="Kyphosis决策树",family='Helvetica',font=5);
```
