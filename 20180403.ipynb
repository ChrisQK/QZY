```julia
function Gradient_descent(f,g,x0;
        ϵg=0.0001,ϵx=0.0001,ϵf=0.0001,maxIteractions=128,verbose=true)
    d = -g(x0...)
    f0 = f(x0...)
    ng = norm(d)
    α = Inexact_method(f,g,f0,-ng'*ng,x0,d)
    x1 = x0+α.*d
    f1 = f(x1...)
    nx = α*ng             ##x1-x0=x0+α*ng-x0；
    nstep = 1
    println("step = ",nstep,"\t","x = ",x1,"\t","f = ",f1,"\t","α = ",α,"\t","g = ",-d)
    while (ng>ϵg || nx>ϵx || abs(f1-f0)>ϵf ) && nstep<maxIteractions  ##当三个同时成立时，循环终止；
        x0 = x1
        d = -g(x0...)
        f0 = f1
        ng = norm(d)
        α = Inexact_method(f,g,f0,-ng'*ng,x0,d)
        x1 = x0 + α.*d
        f1 = f(x1...)
        nx = α*ng                  
        nstep+=1
        if verbose
            println("step = ",nstep,"\t","x = ",x1,"\t","f = ",f1,"\t","g = ",-d)
            println("\t","d = ",d,"\t","α = ",α)
        end
    end
    if nstep>=maxIteractions
        println("warning:Iteractions exceeds",maxIteractions)
    end
    return x1,f1
end
```

```julia
function Inexact_method(f,g,ϕ0,ϕd0,xk,dk;  ##原函数、一阶导、上一点值、α函数一阶为零值、初始点、负梯度（反方向）；
        τ=0.5,ϵ=0.5,ζ=2)
    α=1
    dα=dk'*g((xk.+α.*dk)...)   ##一阶导数；  
    while dα<0                  ##判断α是否足够大；
        α=ζ*α
        dα=dk'*g((xk.+α.*dk)...)
    end
    ϕα=f((xk.+α.*dk)...)      ##α的原函数值；
    while(ϕα>ϕ0+ϵ*α*ϕd0)     ##判断α是否不会太大；
        α=τ*α
        ϕα=f((xk.+α.*dk)...)
    end
    return α
end
```

```julia
function gradient_descent(f,g,x0;
        ϵg=0.0001,ϵx=0.0001,ϵf=0.0001,maxIteractions=128,verbose=true)
    d = -g(x0)
    f0 = f(x0)
    ng = norm(d)
    α = inexact_method(f,g,f0,-ng'*ng,x0,d)
    x1 = x0+α.*d
    f1 = f(x1)
    nx = α*ng
    nstep = 1
    println("step = ",nstep,"\t","x = ",x1,"\t","f = ",f1,"\t","α = ",α,"\t","g = ",-d)
    while (ng>ϵg || nx>ϵx || abs(f1-f0)>ϵf ) && nstep<maxIteractions
        x0 = x1
        d = -g(x0)
        f0 = f1
        ng = norm(d)
        α = inexact_method(f,g,f0,-ng'*ng,x0,d)
        x1 = x0 + α.*d
        f1 = f(x1)
        nx = α*ng                  
        nstep+=1
        if verbose
            println("step = ",nstep,"\t","x = ",x1,"\t","f = ",f1,"\t","g = ",-d)
            println("\t","d = ",d,"\t","α = ",α)
        end
    end
    if nstep>=maxIteractions
        println("warning:Iteractions exceeds",maxIteractions)
    end
    return x1,f1
end
```

```julia
function inexact_method(f,g,ϕ0,ϕd0,xk,dk;
        τ=0.5,ϵ=0.5,ζ=2)
    α=1
    dα=dk'*g((xk.+α.*dk))
    while dα<0
        α=ζ*α
        dα=dk'*g((xk.+α.*dk))
    end
    ϕα=f((xk.+α.*dk))
    while(ϕα>ϕ0+ϵ*α*ϕd0)
        α=τ*α
        ϕα=f((xk.+α.*dk))
    end
    return α
end
```
